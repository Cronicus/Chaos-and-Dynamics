```python
%matplotlib notebook
import numpy as np
from numpy import linalg as lin
from matplotlib import pyplot as plt
from matplotlib.animation import FuncAnimation
tp=10.0**(-3)
def acc(pos1,pos2,m2):
    accx=-39.748*m2*(pos1[0]-pos2[0])/(lin.norm(pos1-pos2))**(3)
    accy=-39.748*m2*(pos1[1]-pos2[1])/(lin.norm(pos1-pos2))**(3)
    l=np.array([accx,accy])
    return l
def energy(pos1,pos2,v):
    eng=(v[0]**2+v[1]**2)/2 -39.748/lin.norm(pos1-pos2)
    return eng
p1=np.array([1.5,-10.0])
p2=np.array([1.0,0.0])
p3=np.array([-1.0,0.0])
v1=np.array([0.0,2.0])
v2=np.array([0.0,-3.0])
v3=np.array([0.0,3.0])
tp=10.0**(-3)
l1x=[];l2x=[];l3x=[];
l1y=[];l2y=[];l3y=[];
for i in range(0,10000):
    l1x.append(p1[0])
    l2x.append(p2[0])
    l3x.append(p3[0])
    l1y.append(p1[1])
    l2y.append(p2[1])
    l3y.append(p3[1])
    p1+=v1*tp
    p2+=v2*tp
    p3+=v3*tp
    v1+=(acc(p1,p2,1.0)+acc(p1,p3,1.0))*tp
    v2+=(acc(p2,p1,1.0)+acc(p2,p3,1.0))*tp
    v3+=(acc(p3,p1,1.0)+acc(p3,p2,1.0))*tp
```

```python
fig,ax=plt.subplots()
ax.set_xlim(-6.0,6.0)
ax.set_ylim(-6.0,6.0)
line1,=ax.plot(0,0)
line2,=ax.plot(0,0)
line3,=ax.plot(0,0)
def anime(i):
    line1.set_xdata(l1x[:i])
    line1.set_ydata(l1y[:i])
    line2.set_xdata(l2x[:i])
    line2.set_ydata(l2y[:i])
    line3.set_xdata(l3x[:i])
    line3.set_ydata(l3y[:i])
  
    return line1,line2,line3,
function=FuncAnimation(fig,func=anime,frames=10000,interval=1,repeat=False)
plt.show()   
```
